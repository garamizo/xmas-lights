```python
import numpy as np
from scipy.io.wavfile import read, write 
import matplotlib.pyplot as plt 
from numpy.fft import fft, ifft 

Fs, data = read('data/smash.wav')
```

```python
# plt.plot(data[int(1*Fs):(2*Fs),:]);
from scipy import signal

sig = data[1*Fs : 10*Fs,0]
# time = np.arange(0, 10, 1/Fs)
# sig = np.sin(2*np.pi*5000*time) + np.random.randn(time.shape[0])

f, t, Zxx = signal.stft(sig, Fs, nperseg=0.2*Fs)
plt.pcolormesh(t, f, np.log(np.abs(Zxx)), vmin=0, vmax=100, shading='gouraud');
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]');
```

```python
Fs_new = 10
N = int(Fs / Fs_new)

sig = data[1*Fs : 10*Fs,0]

val = []
for i in range(sig.shape[0] / N):
    if (i+1)*N >= sig.shape[0]:
        break 
    ampl = sig[i*N : (i+1)*N].max() - sig[i*N : (i+1)*N].min()
    val.append(ampl)

plt.plot(val);
```
